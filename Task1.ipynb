```python
import matplotlib.pyplot as plt
import pandas as pd

file_path = 'cities_r2.csv'
df = pd.read_csv(file_path)
```

```python
# Bar Chart: Distribution of Cities by State
state_counts = df['state_name'].value_counts()

plt.figure(figsize=(12, 6))
state_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Cities by State')
plt.xlabel('State')
plt.ylabel('Number of Cities')
plt.xticks(rotation=90)
plt.show()
```

```python
# Histogram: Distribution of Total Population
plt.figure(figsize=(12, 6))
plt.hist(df['population_total'], bins=30, color='green', edgecolor='black')
plt.title('Distribution of Total Population')
plt.xlabel('Total Population')
plt.ylabel('Number of Cities')
plt.show()
```
